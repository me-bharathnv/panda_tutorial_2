```python
# adding values to new columns
import pandas as pd
data = pd.read_excel("amazon.xlsx")
data["Total"] = data["Offered_price_in_rupee"] + data["real_rice_in_rupee"]
print(data.head(5))
```

```python
# adding columns using index
data["Total"] = data.iloc[:,4:6].sum(axis = 1)
print(data)
```

```python
data.columns
```

```python
# changing positions of columns
data = data[['phone_name', 'Brand', 'ratting_out_of_five', 'Rated_people',
       'Offered_price_in_rupee', 'real_rice_in_rupee','Total', 'Percent_off']]
print(data.head(5))
```

```python
# dropping columns
import pandas as pd
data = pd.read_excel("amazon.xlsx")
data = data.drop(columns = ["Percent_off"])
print(data)
```

```python
# saving file in different format
import pandas as pd
data = pd.read_excel("amazon.xlsx")
data["Total"] = data["Offered_price_in_rupee"] + data["real_rice_in_rupee"]
data = data[['phone_name', 'Brand', 'ratting_out_of_five', 'Rated_people',
       'Offered_price_in_rupee', 'real_rice_in_rupee','Total', 'Percent_off']]
data.to_csv("amazon_new.csv", index=False)
data.to_excel("amazon_new.xlsx", index = False)
data.to_csv("amzon_new.txt", index = False, sep = "\t")
```

```python
import pandas as pd
data = pd.read_csv("amazon_new.csv")
print(data.head(5))
```

```python
# filter based on conditions
samsung = data.loc[(data["Brand"]== "Samsung") & (data["ratting_out_of_five"] > 4.0)]
samsung = samsung.reset_index() # reseting index
print(samsung)
print(samsung.count())
```

```python
samsung.loc[samsung["phone_name"].str.contains("Blue")] # search text and filter data
```

```python

import re
samsung.loc[samsung["phone_name"].str.contains("Brown" and "Black", flags = re.I, regex = True)]
```

```python
import pandas as pd
data = pd.read_csv("amazon_new.csv")
print(data)
```

```python
# aggrigate statistics
data.groupby(["Brand"]).mean().sort_values("ratting_out_of_five") # group by values for comparision
```

```python
data.groupby(["Brand"]).sum().sort_values("Percent_off") # group by values for comparision
```

```python
data.groupby(["Brand"]).count().sort_values("Percent_off") # group by values for comparision
```
